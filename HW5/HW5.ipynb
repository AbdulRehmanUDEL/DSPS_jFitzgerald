## ***We lost just over 3000 rows (3000 buildings)***

I'm going to use a *two-sample t test* to see if there's a difference between the full and reduced datasets. The general equation for this test is
$$t=\frac{(\bar X_1-\bar X_2)-(\mu_1-\mu_2)}{\sqrt{\frac{\sigma_1^2}{n_1}+\frac{\sigma_2^2}{n_2}}}$$

In this case, I'm taking the null hypothesis that the two samples are the same, so $\mu_1-\mu_2=0$:

$$t=\frac{\bar X_1-\bar X_2}{\sqrt{\frac{\sigma_1^2}{n_1}+\frac{\sigma_2^2}{n_2}}}$$

```python
# i'm comparing llplreduced and llplfilter
for c in ['Year Built','Occupancy','Electricity Use - Grid Purchase (kBtu)', 'numfloors', 'ENERGY STAR Score']:
  set1 = llplfilter[c].apply(pd.to_numeric,errors='coerce').dropna()
  set2 = llplreduced[c].apply(pd.to_numeric,errors='coerce').dropna()
  
  xbar1 = set1.mean()
  xbar2 = set2.mean()
  
#  print('averages',xbar1,xbar2)
  
  var1 = set1.var()
  var2 = set2.var()
  
  n1 = set1.count()
  n2 = set2.count()
#  print(n1, n2)
  t = (xbar1 - xbar2) / (var1/n1 + var2/n2)**0.5
  
  if t > 2 or t < -2:
    condstring = ""
  else:
    condstring = "*in*"
  
  print('The t score for \"{}\" was {:0.2f}, showing a \n\
  statistically {}significant change from {:.4} to {:.4} upon reduction\n\
  of the dataset, at the p<0.05 confidence level.\n'
        .format(c, t, condstring, xbar1, xbar2))
```

It appears that the shorter buildings with higher occupancy tended to have more complete datasets. However, these changes were quite small, and furthermore, the rest of this analysis will depend mostly on electricity use, and the Electricity Use and ENERGY STAR Score means did not change much after eliminating incomplete rows.

```python
sets = np.empty(2,dtype='O')

for c in ['Year Built','Occupancy','Electricity Use - Grid Purchase (kBtu)',
          'numfloors', 'ENERGY STAR Score']:
  sets[0] = llplfilter[c].apply(pd.to_numeric,errors='coerce').dropna()
  sets[1] = llplreduced[c].apply(pd.to_numeric,errors='coerce').dropna()
#   print(len(sets[0]))
  for i in [0,1]:
    for _ in [1,2,3]:
#       print(sets[i].mean())
      sets[i] = sets[i].where(sets[i]/sets[i].mean() > 0.01)/sets[i].mean()
      sets[i] = sets[i].where(sets[i]/sets[i].mean() < 100)/sets[i].mean()
      sets[i] = sets[i].dropna()
      sets[i] = sets[i]/sets[i].mean()
#  print(sets[0].min(),sets[0].mean(),sets[0].max())
#  print(sets[1].min(),sets[1].mean(),sets[1].max(),'\n')
  
#  print('---')

  print(c, '\n ks-statistic: {:0.8f},\n pvalue: {:0.8f}\n'.format(stats.ks_2samp(sets[0],sets[1]).statistic,
                                         stats.ks_2samp(sets[0],sets[1]).pvalue))
```

***So the KS-test (after removing outliers of $\mu\times 100$ and $\mu \div 100$) shows that the only statistically significant change between the datasets is the occupancy of the buildings, which was already shown to be a change only from 98% to 99%.***
