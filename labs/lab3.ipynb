```
def square(length):
  ax = plt.figure(figsize=(7,7)).add_subplot(1,1,1)
  plt.plot([length,length,-length,-length,length],[length,-length,-length,length,length],'-k')
  #this line functions on the last figure that was made
  
  return ax
  
square(2)
```

```
from astropy import units as u

def circle(radius,ax):
  angle = np.arange(0,360,1)
  angle *= u.degree
  x = radius * np.cos(angle.to(u.rad))
  y = radius * np.sin(angle.to(u.rad))
  #print('x:',x)
  #print('y:',y)
  ax.plot(x,y,'-k')
```

```
axis = square(1)
circle(1,axis)
```

```
axis = square(1)
circle(1,axis)

np.random.seed(10812355)

numpoints = 100000

xrand = np.random.rand(numpoints) * 2 - 1
yrand = np.random.rand(numpoints) * 2 - 1

axis.plot(xrand,yrand,',')

r = np.sqrt((xrand**2+yrand**2)**0.5)

#print(r)

r_in = r[r < 1]
print(len(r_in)/len(r)*4)

```
